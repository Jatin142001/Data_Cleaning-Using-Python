# Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

# Feature Scaling using Normalization Technique(Min_Max_Technique)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_ms"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

# Showing the histogram plot of the Scaling 

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CoapplicantIncome_ms"])

plt.show()
```
