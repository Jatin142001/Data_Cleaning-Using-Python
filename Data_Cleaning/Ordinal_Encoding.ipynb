# Import required libraries

```python
import pandas as pd
```

# Creating Dataset

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","s","xl","m"]})
df.head(5)
```

# Giving the Order Priority

```python
ordinal_data = [["s","m","l","xl"]]
```

# Ordinal Encoding through Scikit-learn

```python
from sklearn.preprocessing import OrdinalEncoder
```

# Fit the Ordinal Encoding

```python
ord_en = OrdinalEncoder(categories= ordinal_data)
ord_en.fit(df[["Size"]])
```

# Transform Ordinal Encoding

```python
df["Size_en"]=ord_en.transform(df[["Size"]])
```

```python
df
```

# Ordinal Encoding through Mapping

```python
ord_data1 = {"s":1,"m":2,"l":3,"xl":4}
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
```
