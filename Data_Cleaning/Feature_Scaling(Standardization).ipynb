# Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

# Finding the null values present in the dataset

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns = ["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

# Showing the histogram plot of the feature

```python
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["ApplicantIncome_ss"])
plt.show()
```
